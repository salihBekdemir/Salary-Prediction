```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('Salary_Data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.lineplot(x='YearsExperience',y='Salary', data=df).set_title("Salary in years");
```

```python
x=df.drop('Salary',axis=1)
y=df[['Salary']]
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(230,activation='relu'))
model.add(Dense(120,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1)) 
model.compile(loss='mse', optimizer='adam')
```

```python
history=model.fit(x_train,y_train, validation_data=(x_test,y_test),batch_size=128,epochs=1500,verbose=1)
```

```python
tahmin=model.predict(x_test)
```

```python
r2_score(y_test,tahmin)
```

```python
mean_squared_error(y_test,tahmin)**.5
```

```python
model.save('Salary.keras')
```

```python
import matplotlib.pyplot as plt
```

```python
loss_f=pd.DataFrame(history.history)
```

```python
loss_f.plot();
```
